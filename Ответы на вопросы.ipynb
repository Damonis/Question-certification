1. В задачах машинного обучения для оценки качества моделей и сравнения различных алгоритмов используются метрики

Метрики часто делят на группы в зависимости от модели для которых они используются:

    Метрики классификации

    Метрики регрессии

    Рейтинговые метрики
    
    Статистические метрики

    И т.д

Основных метриков из группы классификации можно выделить 3:

Точность(accuracy)

Полнота(Recall)

Точность положительного класса(precision)

Все эти метрики полагаются на матрицу ошибок(confusion matrix),которая показывает сколько раз модель дала правильное или неправильное предсказание для каждого класса

Из вторичных метриков можно выделить:
F1-мера — это гармоническое среднее между точностью и полнотой.

ROC-кривая— это график зависимости доли верных положительных классов (TPR) от доли ложных положительных классов (FPR) при различных пороговых значениях классификационной функции.

Давайте предположим, что в нашем тестовом наборе 1100 изображений (1000 изображений, отличных от попугаев, и 100 изображений попугаев)

Из 100 изображений попугаев модель предсказала 90 из них правильно и неправильно классифицировала 10 из них.

Из 1000 изображений, не относящихся к попугаям, модель правильно классифицировала 940 из них, а 60 ошибочно классифицировала их.

В таком случае посчитаем точность(precision) модели 

Точность = True_Positive / (True_Positive + False_Positive)

Precision_parrot= # образцы правильно предсказали попугая / # образцы предсказали как попугая = 90 / (90 + 60) = 60%

Precision_NonParrot = 940/950 = 98,9%

2.Что такое тестовый набор данных (test set) и для чего он нужен?

В машинном обучении тестовое множество — это подмножество обучающего набора данных, содержащее тестовые примеры, т.е. примеры, использующиеся не для обучения модели, а для проверки его результатов.

Цель применения тестового множества – проверить, как обученная модель будет работать с новыми данными, т.е. приобрела ли она обобщающую способность.

3.Каково назначение валидационного множества данных (validation set)?

Данные, используемые при обучении программы искусственного интеллекта (ИИ) поиску и оптимизации наилучшей модели для решения заданной задачи, называются validation set. Это также может называться development или dev set. Набор проверок используется для точной настройки модели искусственного интеллекта. Результаты набора проверок обновят гиперпараметры более высокого уровня.

Основная цель validation set состоит в том, чтобы предотвратить переобучение модели. Переобучение - это когда модель запоминает шаблон в обучающих данных, но не может делать точные прогнозы на основе того, что она еще не проанализировала. В результате модель будет просматривать данные, но не извлекать из них уроки.

6.Что такое перекрестная проверка (cross-validation) и чем она лучше
использования одного тестового множества?

Кросс-валидация — это метод, предназначенный для оценки качества работы модели, широко применяемый в машинном обучении. Он помогает сравнить между собой различные модели и выбрать наилучшую для конкретной задачи.То есть,данный метод способствует более точной оценки работы модели за счет использования несколько тестирующих и обучающих множеств

7.Что такое онлайновый режим работы системы машинного обучения?

Онлайн-машинное обучение - это тип машинного обучения, при котором данные собираются последовательно и используются для обновления наилучшего прогноза будущих данных на каждом этапе.

Другими словами, машинное обучение в режиме онлайн означает, что обучение происходит по мере поступления данных. При онлайн-обучении параметры алгоритма обучения обновляются после изучения каждого отдельного обучающего примера. При онлайн-обучении каждый шаг обучения является быстрым и дешевым, и модель может извлекать уроки из новой информации в режиме реального времени по мере ее поступления.

Линейная регрессия является линейной моделью,например модель, которая предполагает линейную связь между входными переменными (x) и единственной выходной переменной (y). Более конкретно, что  можно вычислить из линейной комбинации входных переменных (х).



Различные методы могут быть использованы для подготовки или обучения уравнения линейной регрессии из данных, наиболее распространенный из которых называется обычный наименьший квадрат

Предсказание представляет собой линейное уравнение, объединяющее определенный набор входных значений (x), решением которого является прогнозируемый результат для этого набора входных значений (y). Таким образом, как входные значения (x), так и выходное значение являются числовыми.

10.В чем заключается разница между обучением с учителем и обучением без
привлечения учителя? Какие ещё есть «промежуточные» типы обучения.


Обучение нейросети с учителем. 

Его суть в том, что для обучения нейросеть получает специальный набор данных (датасет), в котором заранее отмечено, что эти данные означают. То есть нейросеть получает и вопрос, и ответ, который должна давать. Она «смотрит» на большие объёмы таких данных и так учится отвечать правильно.Чтобы такое обучение было возможным, нужно предварительно собрать, отсмотреть вручную и разметить обучающий датасет. Этим должен заниматься человек или группа людей — как раз тот самый условный «учитель».

Обучение без учителя.

Его суть в том, что нейросеть получает на входе неразмеченные данные и старается сама найти в них общие признаки и связи. То есть сначала нейросети скармливают очень много разных данных — а потом выпускают работать в «реальный мир», с настоящими задачами и всё с такими же неразмеченными данными.Иногда совсем без разметки нейросеть обучиться неспособна, но много размеченных данных получить нельзя. Тогда используют частичное привлечение учителя — часть данных размечают, а часть передают в сыром виде. Это позволяет ускорить обучение и сделать его более точным. Например, так часто делают с медицинскими данными — передают рентгеновские снимки, на малом числе которых патология обозначена однозначно, а на других нейросеть учится ее распознавать.

Также существует еще один вид обучения, а именно машинное обучение с подкреплением

Суть метода в том, что нейронная сеть получает входные данные и случайным образом их обрабатывает. Потом результат её работы оценивают по определенным критериям. Нейросеть видит эти критерии — и в следующий раз обрабатывает данные так, чтобы им соответствовать. Прежде чем удается сформировать оптимальную стратегию проходит много итераций. Для этого не нужен учитель или специальный обучающий набор данных — только чёткие задачи и критерии оценки их выполнения.

11.Что такое переобучение?


Переобучение – это явление, когда качество модели на обучающей выборке существенно превосходит качество модели на тестовой выборке.

Основной метод борьбы с переобучением это регуляция:

Регуляризация, относится к минимизации или уменьшению оценок коэффициентов до нуля, чтобы избежать недостаточной или чрезмерной подгонки модели машинного обучения.
Разница заключается в том, как мы обращаем внимание на данные и модель машинного обучения. Эти многотомные тома о моделях машинного обучения (в частности, линейной регрессии) также будут включать "коэффициенты" для входных параметров. Это машинный эквивалент внимания или важности, приписываемых каждому параметру. По сути, чем выше коэффициент входного параметра, тем более критичным модель машинного обучения считает этот параметр.

12.Опишите кратко метод работы дерева решений и ансамбля таких деревьев
в режиме взаимного усиления (мы знаем его как «бустинг»)

Основная идея любого алгоритма дерева решений заключается в следующем:

1.Выберите лучший атрибут, используя Меры выбора атрибутов (ASM), чтобы разделить записи.

2.Сделайте этот атрибут узлом решения и разбейте набор данных на меньшие подмножества.

3.Начинает построение дерева, повторяя этот процесс рекурсивно для каждого потомка, пока не будет выполнено одно из условий:

o Все кортежи принадлежат одному и тому же значению атрибута.

o Больше нет оставшихся атрибутов.

o Больше нет экземпляров.

Бустинг – это метод, используемый в машинном обучении для уменьшения количества ошибок при прогностическом анализе данных.

Бустинг строит ансамбли моделей путем последовательного объединения нескольких слабых деревьев решений. Выходным данным отдельных деревьев присваиваются веса. Затем неправильным классификациям из первого дерева решений присваивается больший вес, после чего данные передаются в следующее дерево. После многочисленных циклов бустинг объединяет слабые классификаторы в один мощный алгоритм прогнозирования.
